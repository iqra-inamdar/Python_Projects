```python
import pandas as pd  
# pandas (pd): Used for data loading, cleaning, and analysis 
# Example: reading CSV, filtering rows, handling missing values


import plotly.express as px  
# plotly.express (px): Used for quick and easy data visualizations
# Example: bar charts, line charts, pie charts, scatter plots


import plotly.graph_objects as go  
# plotly.graph_objects (go): Used for advanced and customizable charts
# Example: complex layouts, detailed control over graph design


import plotly.io as pio  
# plotly.io (pio): Used to manage templates and output formats for Plotly charts
# Example: setting a default theme for all plots


import plotly.colors as colors  
# plotly.colors: Provides color palettes and tools to manage colors in Plotly charts
# Example: using predefined color sets or custom color schemes

# Setting Default Plotly Theme

pio.templates.default = "plotly_white"
# This sets the default template for all graphs to "plotly_white"
# It gives a clean, professional white background to every chart automatically
```

### converting Date columns

```python
data['Order Date'] = pd.to_datetime(data['Order Date'])
data['Ship Date'] = pd.to_datetime(data['Ship Date'])
```

```python
data.info()
```

### Adding new Date-Based Columns

```python
data['Order Month'] = data['Order Date'].dt.month
data['Order year'] = data['Order Date'].dt.year
data['Order Day of week'] = data['Order Date'].dt.dayofweek
```

```python
data.head()
```

### Sales by Months

```python
Sales_by_month =data.groupby('Order Month')['Sales'].sum().reset_index()
Sales_by_month
```

```python
fig =  px.line(Sales_by_month,
              x = 'Order Month',
              y = 'Sales',
              title = 'Total Sales by Months'
              )
fig.show()
```

### Sales by Category

```python

Sales_by_Category = data.groupby('Category')['Sales'].sum().reset_index()
Sales_by_Category
```

```python

fig = px.pie(Sales_by_Category,
             values='Sales',
             names='Category',
             hole=0.5,
             color_discrete_sequence=px.colors.qualitative.Pastel

            )
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.update_layout(title_text='Sales Analysis by Category',title_font=dict(size=25))

fig.show()

```

### Sales analysis by sub-category

```python
Sales_by_SubCategory = data.groupby('Sub-Category')['Sales'].sum().reset_index()
Sales_by_SubCategory
```

```python
fig = px.bar(Sales_by_SubCategory,
             x = 'Sub-Category',
             y = 'Sales',
             title = 'Sales analysis by sub Category')

fig.show()
```

```python
fig = px.line(profit_by_month,
              x = 'Order Month',
              y = 'Profit',
              title = 'Monthly profit Analaysis')
fig.show()
```

### Sales and Profit by customer segment 

```python
sales_profit_by_segment = data.groupby('Segment').agg({'Sales': 'sum', 'Profit':'sum'}).reset_index()
sales_profit_by_segment
```

```python
color_palette = colors.qualitative.Pastel
fig = go.Figure()
fig.add_trace(go.Bar(x=sales_profit_by_segment['Segment'],
                     y=sales_profit_by_segment['Sales'],
                     name = 'Sales',
                     marker_color=color_palette[0]))
fig.add_trace(go.Bar(x=sales_profit_by_segment['Segment'],
                     y=sales_profit_by_segment['Profit'],
                     name = 'Profit',
                     marker_color=color_palette[1]))
fig.update_layout(title='Sales and Profit Analysis by Customer Segment',
                  xaxis_title='Customer Segment',yaxis_title='Amount')
fig.show()

```

### Sales to Profit ratio

```python
sales_profit_by_segment=data.groupby('Segment').agg({'Sales':'sum','Profit':'sum'}).reset_index()
sales_profit_by_segment['sales_by_profit_ratio']=sales_profit_by_segment['Sales']/sales_profit_by_segment['Profit']
sales_profit_by_segment[['Segment','sales_by_profit_ratio']]
```
